```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import optimize
import pandas as pd
import seaborn as sns

import scienceplots  
plt.style.use(['science','notebook', 'grid'])

from matplotlib import rc
rc('text', usetex=True)
font = {'family' : 'serif',
        'weight' : 'bold',
        'size'   : '14'}
rc('font', **font)
```

### Graph of $\ln{x}$ vs time 


```python
df = pd.read_excel('ln table.xlsx')

data_array = df.to_numpy()
transposed_array = data_array.transpose()
x = np.array(transposed_array[0], dtype='float')
y = np.array(transposed_array[1], dtype='float')
```

```python
print(x)
print(y)
```

```python
coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef) 
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x,y, 'bo', x, poly1d_fn(x), '--k') #'--k'=black dashed line, 'yo' = yellow circle marker
ax.set_xlabel(r'Time [s]')
ax.set_ylabel(r'$\ln{x}$')
plt.text(0.35, 0.85, '$y = (-0.0257 \pm 0.0005)x + (-2.548 \pm 0.003)$', transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='red'))
# plt.savefig('natural_log_vs_time.png')
plt.show()
```

### Theoretical & Experimental Position vs Time graph

```python
df1 = pd.read_excel('Residual.xlsx')

data_array1 = df1.to_numpy()
transposed_array1 = data_array1.transpose()

x1 = np.array(transposed_array1[0])
# theoretical value
y1 = np.array(transposed_array1[1]) 
# experimental
y2 = np.array(transposed_array1[2])
```

```python
fig, ax = plt.subplots(nrows=3, ncols=1)
ax[0].plot(x1, y1, color='blue')
ax[1].plot(x1, y2, color='red')
ax[2].scatter(x1, y1, c='b', marker='.')
# ax[2].scatter(x1, y2, c='r', marker='x')
ax[2].plot(x1, y2, c='green')
ax[0].set_ylabel('Theoretical')
ax[1].set_ylabel('Experimental')
ax[2].set_ylabel('Theo vs Exp')
fig.supxlabel('Time (s)', size='x-large')
fig.supylabel('Position (m)', size='x-large', ha='right')
plt.show()
```

```python
plt.plot(x1, y1, c='red')
plt.plot(x1, y2, c='blue')
plt.scatter(x1, y1, c='r', marker='+', label='Theoretical')
plt.scatter(x1, y2, c='b', marker='+', label='Experimental')
plt.xlabel('Time[s]')
plt.ylabel('Position (m)')
plt.legend()
plt.savefig('theo_vs_exp.png')
plt.show()
```

```python
# just for comparison's sake
fig, ax = plt.subplots(nrows=2, ncols=1)

ax[0].plot(x1, y2, c='green', label='experimental')
ax[0].plot(x1, y1, c='pink', label='theoretical')
# plt.scatter(x1, y2, c='r', marker='+', label='Experimental')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
ax[0].legend(loc=(1.02, 0.6))

ax[1].plot(x1, y1, c='red', label='theoretical')
ax[1].plot(x1, y2, c='green', label='experimental')
plt.scatter(x1, y1, c='r', marker='+', label='Theoretical')
plt.scatter(x1, y2, c='b', marker='+', label='Experimental')
ax[1].legend(loc=(1.02, 0.3))

plt.show()
```
